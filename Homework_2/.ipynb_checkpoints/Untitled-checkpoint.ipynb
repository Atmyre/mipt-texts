Посмотрим, сколько различных значений DocNumber, MetadataPdfLink, MetadataCaseNumber встречается в датасете:

```python
print("Different values of:")

print("DocNumber: ", len(pd.unique(emails.DocNumber)))
print("MetadataPdfLink: ", len(pd.unique(emails.MetadataPdfLink)))
print("MetadataCaseNumber: ", len(pd.unique(emails.MetadataCaseNumber)))
```

Ясно, что признаки DocNumber и MetadataPdfLink неинформативны, уберем их:

```python
for feature in ['DocNumber', 'MetadataPdfLink']:
    emails.pop(feature)
```

Также удалим MetadataFrom, так как есть признак SenderId, который кажется более информативным:

```python
emails.pop('MetadataFrom')
```

Также удалим ExtractedTo, ExtractedFrom, ExtractedCc, ExtractedDateSent, ExtractedCaseNumber, ExtractedDocNumber, ExtractedDateReleased, ExtractedReleaseInPartOrFull, так как, во-первых, большинство из этой информации есть в признаках Metadata\*, а во-вторых, в этих данных много пропусков (около половины значений --- NaN).
Здесь же удалим RawText, так как есть ExtractedBodyText:

```python
for feature in ['ExtractedTo', 'ExtractedFrom', 'ExtractedCc', 'ExtractedDateSent', 'ExtractedCaseNumber', 
                'ExtractedDocNumber', 'ExtractedDateReleased', 'ExtractedReleaseInPartOrFull', 'ExtractedSubject', 
                'RawText']:
    emails.pop(feature)
```

```python
emails.head()
```

Посмотрим на уникальные значения признака MetadataTo:

```python
pd.unique(emails.MetadataTo)
```

Уберем все окончания после @ и заменим ;H на H:

```python
import re

for i in range(len(emails.MetadataTo)):
    if emails.MetadataTo[i] == ';H':
        emails.MetadataTo[i] = 'H'
    if len(re.findall(r'@', str(emails.MetadataTo[i]))) == 0:
        continue
        
    emails.MetadataTo[i] = re.findall(r'(.*)@', str(emails.MetadataTo[i]))[0]
```

Переведем теперь все строковые данные (кроме ExtractedBodyText) в числовые и заполним NaN:

```python
def make_data_numeric(column):
    value_to_num = dict()
    
    for i in range(len(column)):
        if str(column[i]) not in value_to_num:
            value_to_num[str(column[i])] = len(value_to_num)
        column[i] = value_to_num[str(column[i])]
```

```python
for column in ['MetadataSubject', 'MetadataTo', 'SenderPersonId', 'MetadataDateSent', 'MetadataDateReleased',
               'MetadataCaseNumber', 'MetadataDocumentClass']:
    make_data_numeric(emails[column])
```

Также уберем некторую лишнюю информацию из ExtractedBodyText, а те строчки, где ExtractedBodyText = NaN, удалим (даже не extracted эти письма были странными, может, пустыми вообще):

```python
# удаляем строчки с ExtractedBodyText = NaN
emails = emails[emails.ExtractedBodyText.notnull()]
```

```python
patterns_to_delete = ['Subject[:]*', 'Case', 'From', 'To', 'UNCLASSIFIED', 'Sent:', 'Date', 'Doc', 'Cc',  'Re',
                      'U.S. Department', 'STATE DEPT', 'SUBJECT TO', 'RELEASE', '\x0c', 'Fw:', 'cc', '\<(.*?)\>', 
                      'H', '[\n]+', 'No. [A-Z][0-9,-]*', ':{\s}{>0}\n', ',{\s}{>1}', '{\s}{>1}', '{\n}{>1}', 'Cc'
                      # time
                      '[0-9]{1,2}, 20[0-9]{2} [0-9]{1,2}:[0-9]{2} AM',
                      '[0-9]{1,2}, 20[0-9]{2} [0-9]{1,2}:[0-9]{2} PM', 
                       # months
                      'January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 
                      'October','November', 'December', 
                      # day of week
                      'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
```

```python
import re
from tqdm import tqdm

for i in tqdm(emails.axes[0]):
    for to_delete in patterns_to_delete:
        emails.ExtractedBodyText[i] = re.sub(to_delete, '', str(emails.ExtractedBodyText[i]))
```
